```python
import numpy as np
import matplotlib.pyplot as plt

def silnia(n):
    if n == 0:
        return 1
    else:
        return n * silnia(n - 1)

def tsin(x, n):
    sin = 0
    for i in range(n):
        sin = sin + ((-1)**i / silnia(2 * i + 1)) * (x**(2 * i + 1))
    return sin

def tcos(x, n):
    cos = 0
    for i in range(n):
        cos = cos + ((-1)**i / silnia(2 * i)) * (x**(2 * i))
    return cos

def texp(x, n):
    exp = 0
    for i in range(n):
        exp = exp + (1 / silnia(i)) * x**i
    return exp

def ttan(x, n):
    sin = tsin(x, n)
    cos = tcos(x, n)
    tan = sin / cos
    return tan

def errors_sin(xp,xk):
    errors = []
    for i in np.arange(xp, xk, 0.1):
        bezwzgledny = abs(np.sin(i) - tsin(i, 10))
        wzgledny = abs(np.sin(i) - tsin(i, 10)) / np.sin(i)
        #print("np.sin = " + str(np.sin(i))+ " tsin = " + str(tsin(i, 10)))
        #print("Blad bezwzgledny = " + str(bezwzgledny))
        #print("Blad wzgledny = " + str(wzgledny))
        errors.append(bezwzgledny)
    odchylenie = np.std(errors)
    mean = np.mean(errors)
    wspolczynnik = odchylenie / mean * 100 
    print("Odchylenie standardowe = "+str(odchylenie))
    print("Wspolcznnik zmiennosci bledow = "+str(wspolczynnik))
    x = [i for i in np.arange(xp,xk,0.1)]
    plt.plot(x,errors)
    plt.ylabel('errors_sin')
    plt.xlabel('x')
    plt.show()
    

errors_sin(-3*np.pi,0)
#errors_sin(0, np.pi/4)
errors_sin(-3*np.pi,3*np.pi)

def errors_cos(xp,xk):
    errors = []
    for i in np.arange(xp, xk, 0.1):
        bezwzgledny = abs(np.cos(i) - tcos(i, 10))
        wzgledny = abs(np.cos(i) - tcos(i, 10)) / np.cos(i)
        #print("np.cos = " + str(np.cos(i))+ " tcos = " + str(tcos(i, 10)))
        #print("Blad bezwzgledny = " + str(bezwzgledny))
        #print("Blad wzgledny = " + str(wzgledny))
        errors.append(bezwzgledny)
    odchylenie = np.std(errors)
    mean = np.mean(errors)
    wspolczynnik = odchylenie / mean * 100
    print("Odchylenie standardowe = "+str(odchylenie))
    print("Wspolcznnik zmiennosci bledow = "+str(wspolczynnik))
    x = [i for i in np.arange(xp,xk,0.1)]
    plt.plot(x,errors)
    plt.ylabel('errors_cos')
    plt.xlabel('x')
    plt.show()
    
errors_cos(-3*np.pi,0)
#errors_cos(0, np.pi/4)
errors_cos(-3*np.pi,3*np.pi)

def errors_exp(xp,xk):
    errors = []
    for i in np.arange(xp, xk, 0.1):
        bezwzgledny = abs(np.exp(i) - texp(i, 10))
        wzgledny = abs(np.exp(i) - texp(i, 10)) / np.exp(i)
        #print("np.exp = " + str(np.exp(i))+ " texp = " + str(texp(i, 10)))
        #print("Blad bezwzgledny = " + str(bezwzgledny))
        #print("Blad wzgledny = " + str(wzgledny))
        errors.append(bezwzgledny)
    odchylenie = np.std(errors)
    mean = np.mean(errors)
    wspolczynnik = odchylenie / mean * 100
    print("Odchylenie standardowe = "+str(odchylenie))
    print("Wspolcznnik zmiennosci bledow = "+str(wspolczynnik))
    x = [i for i in np.arange(xp,xk,0.1)]
    plt.plot(x,errors)
    plt.ylabel('errors_exp')
    plt.xlabel('x')
    plt.show()
    
errors_exp(0,20)
errors_exp(0,1)

def errors_tan(xp,xk):
    errors = []
    for i in np.arange(xp, xk, 0.1):
        bezwzgledny = abs(np.tan(i) - ttan(i, 10))
        wzgledny = abs(np.tan(i) - ttan(i, 10)) / np.tan(i)
        #print("np.tan = " + str(np.tan(i))+ " ttan = " + str(ttan(i, 10)))
        #print("Blad bezwzgledny = " + str(bezwzgledny))
        #print("Blad wzgledny = " + str(wzgledny))
        errors.append(bezwzgledny)
    odchylenie = np.std(errors)
    mean = np.mean(errors)
    wspolczynnik = odchylenie / mean * 100
    print("Odchylenie standardowe = "+str(odchylenie))
    print("Wspolcznnik zmiennosci bledow = "+str(wspolczynnik))
    x = [i for i in np.arange(xp,xk,0.1)]
    plt.plot(x,errors)
    plt.ylabel('errors_tan')
    plt.xlabel('x')
    plt.show()
    
errors_tan(-np.pi,2)
errors_tan(0,np.pi/2)
errors_tan(-np.pi/2,np.pi/2)

# wnioski: 
# wraz ze wzrostem zakresu argumentow rosnie blad wzgledny i bezwzgledny
# wraz ze wzrostem zakresu argumentow rosnie odchylenie standardowe i wspolczynnik zmiennosci bledow
# przy obnizaniu liczby iteracji dla szeregow Taylora blad wzgledny i bezwzgledny rosnie
# dla wartosci bliskich zera blad wzgledny i bezwzgledny jest bliski zeru
# przy obnizaniu liczby iteracji dla szeregow Taylora odchylenie standardowe i wspolczynnik zmiennosci bledow maleje
# przy zwiekszaniu kroku blad wzgledny i bezwzgledny rosnie
# dla sinusa oraz cosinusa blisko wartosci 10 oraz -10 blad bezwzgledny znacznie rosnie
# dla tangensa blad rosnie blisko wartosci sqrt(2) oraz -sqrt(2)
```
