```python
#Zad.1
import numpy as np
import matplotlib.pyplot as plt
def fx1(x):
    return 7*x**5 - 6*x**2 - 2*x

def fx2(x):
    return 1/((x-0.3)**2+0.01) - 1/((x-0.8)**2+0.04)

def fx3(x):
    return x**4 - 6.4*x**3 + 6.45*x**2 + 20.538*x - 31.752

def graph(fx,xstart,xend,step):
    x=np.arange(xstart,xend,step)
    plt.plot(x,fx(x),'r-')
    plt.grid(True)
    plt.show()

print("Metoda graficzna")
graph(fx1,-1,1.5,0.05)
graph(fx2,-1,3,0.05)
graph(fx3,-3,5,0.05)

#

def linear_incrementation(fx,xstart,xd,maxincr):
    x=xstart
    fstart=fx(x)
    for i in range (maxincr):
        x=xstart+i*xd
        if fstart*fx(x)<0:
            break
    if fstart*fx(x)>0:
        raise Exception("Nie znaleziono rozwiazania")
    else:
        return x-(xd*fx(x))/(fx(x)-fx(x-xd))

print("Metoda liniowej inkrementacji")

print("Funkcja 1:")
x1f1 = linear_incrementation(fx1,-0.5,0.01,500)
err_x1f1 = fx1(x1f1)
print("x1=" +str(x1f1))
print("f(x1)=" +str(err_x1f1))
x2f1 = linear_incrementation(fx1,-0.25,0.01,500)
err_x2f1 = fx1(x2f1)
print("x2=" +str(x2f1))
print("f(x2)=" +str(err_x2f1))
x3f1 = linear_incrementation(fx1,1,0.01,500)
err_x3f1 = fx1(x3f1)
print("x3=" +str(x3f1))
print("f(x3)=" +str(err_x3f1))

print("Funkcja 2:")
x1f2 = linear_incrementation(fx2,0.5,0.01,500)
err_x1f2 = fx2(x1f2)
print("x1=" +str(x1f2))
print("f(x1)=" +str(err_x1f2))

print("Funkcja 3:")
x1f3 = linear_incrementation(fx3,-2,0.01,100)
err_x1f3 = fx3(x1f3)    
print("x1=" +str(x1f3))
print("f(x1)=" +str(err_x1f3))
x2f3 = linear_incrementation(fx3,1,0.01,500)
err_x2f3 = fx3(x2f3)
print("x2=" +str(x2f3))
print("f(x2)=" +str(err_x2f3))
x3f3 = linear_incrementation(fx3,3.5,0.01,500)
err_x3f3 = fx3(x3f3)
print("x3=" +str(x3f3))
print("f(x3)=" +str(err_x3f3))


def bisection(fx,a,b,err):
    while np.absolute(b-a)>err:
        midPoint=(a+b)/2
        if fx(midPoint)*fx(a)<0:
            b=midPoint
        midPoint=(a+b)/2
        if fx(midPoint)*fx(b)<0:
            a=midPoint
        return b - (b-a)*fx(b)/(fx(b)-fx(a))
    
print("Metoda bisekcji")
print("Funkcja 1:")
x1f1 = bisection(fx1,-0.5,0.5,0.000001)
errb_x1f1 = fx1(x1f1)
print("x1=" +str(x1f1))
print("f(x1)=" +str(errb_x1f1))
x2f1 = bisection(fx1,-0.25,0.25,0.000001)
errb_x2f1 = fx1(x2f1)
print("x2=" +str(x2f1))
print("f(x2)=" +str(errb_x2f1))
x3f1 = bisection(fx1,1,1.5,0.000001)
errb_x3f1 = fx1(x3f1)
print("x3=" +str(x3f1))
print("f(x3)=" +str(errb_x3f1))

print("Funkcja 2:")
x1f2 = bisection(fx2,0.5,1,0.000001)
errb_x1f2 = fx2(x1f2)
print("x1=" +str(x1f2))
print("f(x1)=" +str(errb_x1f2))

print("Funkcja 3:")
x1f3 = bisection(fx3,-2,0,0.000001)
errb_x1f3 = fx3(x1f3)
print("x1=" +str(x1f3))
print("f(x1)=" +str(errb_x1f3))
x2f3 = bisection(fx3,1,2,0.000001)
errb_x2f3 = fx3(x2f3)
print("x2=" +str(x2f3))
print("f(x2)=" +str(errb_x2f3))
x3f3 = bisection(fx3,3,4,0.000001)
errb_x3f3 = fx3(x3f3)
print("x3=" +str(x3f3))
print("f(x3)=" +str(errb_x3f3))

#Zad.2
print("Najmniejsze miejsce zerowe funkcji 3")
print(str(bisection(fx3,-2,-1,0.000001)))

# Wnioski
# przy doborze mniejszego kroku miejsce zerowe otrzymane 
# metod liniowej inkrementacji jest blisze prawdziwemu
# miejscu zerowemu
# z laboratiorium wynika, e dla zadanych funkcji metoda bisekcji
# daje mniej dokadne wyniki miejsc zerowych ni metoda liniowej inkrementacji
# jednake w przypadku funkcji 3 metoda bisekcji daje dokadniejszy wynik

# Z metody graficznej wynika, e najmniejsze miejsce zerowe funkcji 3 
# znajduje si w okolicach -1.5
# obliczono je zatem metod bisekcji midzy -2 a -1


```
