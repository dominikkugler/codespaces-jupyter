```python
import numpy as np
import matplotlib.pyplot as plt

def linear_approximation(x, y):
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    delta_x = x - mean_x
    delta_y = y - mean_y
    sum_delta_xy = np.sum(delta_x * delta_y)
    sum_delta_x_squared = np.sum(delta_x ** 2)
    b = sum_delta_xy / sum_delta_x_squared
    a = mean_y - b * mean_x
    return a, b
def absolute_error(y_true, y_approx):
    return np.abs(y_true - y_approx)

def relative_error(y_true, y_approx):
    return np.abs((y_true - y_approx) / y_true)

x = np.arange(-20, 20)
y = 0.5 * x**3 + 1.5 * x**2 - 2.5 * x + 69 + np.random.normal(scale=200, size=len(x))

a, b = linear_approximation(x, y)

approximated_y = a + b * x

plt.scatter(x, y, label='dane',marker='o', color='blue')
plt.plot(x, approximated_y, color='red', label='aproksymacja liniowa')
plt.xlabel('x')
plt.ylabel('y')
plt.title('aproksymacja liniowa')
plt.legend()
plt.show()

true_values = 0.5 * x**3 + 1.5 * x**2 - 2.5 * x + 69

absolute_err = absolute_error(true_values, approximated_y)

relative_err = relative_error(true_values, approximated_y)

print("Błąd bezwzględny:")
print(absolute_err)

print("\nBłąd względny:")
print(relative_err)


#Wnioski:
#Aproksymacja liniowa jest dobra dla danych, które są liniowe, ale dla danych nieliniowych jest zła
#Błąd bezwzględny jest największy dla wartości skrajnych
#Błąd bezwzględny zbliża się do zera dla wartości bliskich zeru


```
