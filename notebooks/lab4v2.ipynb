```python
import numpy as np
import matplotlib.pyplot as plt

def interplin2p(x,xi,yi,xi1,yi1):
    return yi+(x-xi)*(yi1-yi)/(xi1-xi)

def interlinevect(x, xyvect):
    xinterp = []
    yinterp = []
    for xk in x:
        N = len(xyvect[0])
        for i in range(0, N - 1):
            if xk >= xyvect[0][i] and xk <= xyvect[0][i + 1]:
                xinterp.append(xk)
                yinterp.append(interplin2p(xk, xyvect[0][i], xyvect[1][i], xyvect[0][i + 1], xyvect[1][i + 1]))
                break
        else:
            xinterp.append(xk)
            yinterp.append(interplin2p(xk, xyvect[0][-2], xyvect[1][-2], xyvect[0][-1], xyvect[1][-1]))

    return [xinterp, yinterp]



x_data = np.arange(0.1, 2.1, 0.1)
y_data1 = np.cos(1 / x_data)
y_data2 = np.exp(np.sin(x_data))

x_interp = np.arange(0.1, 2.1, 0.08)

y_interp1 = interlinevect(x_interp, [x_data, y_data1])[1]
y_interp2 = interlinevect(x_interp, [x_data, y_data2])[1]

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(x_data, y_data1, 'o-', label='Oryginalny wykres funkcji')
plt.title('cos(1/x) - Originalny wykres funkcji')

plt.subplot(2, 2, 2)
plt.plot(x_interp, y_interp1, 's-', label='Interpolowany wykres funkcji')
plt.title('cos(1/x) - Interpolacja liniowa')

plt.subplot(2, 2, 3)
plt.plot(x_data, y_data2, 'o-', label='Oryginalny wykres funkcji')
plt.title('exp(sin(x)) - Originalny wykres funkcji')

plt.subplot(2, 2, 4)
plt.plot(x_interp, y_interp2, 's-', label='Interpolowany wykres funkcji')
plt.title('exp(sin(x)) - Interpolacja liniowa')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def polynomial_interpolation(x_values, y_values, x_interp, degree):
    coefficients = np.polyfit(x_values, y_values, degree)
    y_interp = np.polyval(coefficients, x_interp)
    return y_interp

x_data = np.arange(-np.pi, np.pi, 0.1)
y_data1 = np.sin(1 / x_data)
y_data2 = np.exp(np.cos(x_data))

x_interp = np.arange(-np.pi, np.pi, 0.08)

degree = 5

y_interp1 = polynomial_interpolation(x_data, y_data1, x_interp, degree)
y_interp2 = polynomial_interpolation(x_data, y_data2, x_interp, degree)

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.plot(x_data, y_data1, 'o-', label='Oryginalny wykres funkcji')
plt.title('sin(1/x) - Originalny wykres funkcji')

plt.subplot(2, 2, 2)
plt.plot(x_interp, y_interp1, 's-', label='Interpolowany wykres funkcji')
plt.title('sin(1/x) - Interpolacja wielomianowa (Stopień {})'.format(degree))

plt.subplot(2, 2, 3)
plt.plot(x_data, y_data2, 'o-', label='Oryginalny wykres funkcji')
plt.title('exp(cos(x)) - Originalny wykres funkcji')

plt.subplot(2, 2, 4)
plt.plot(x_interp, y_interp2, 's-', label='Interpolowany wykres funkcji')
plt.title('exp(cos(x)) - Interpolacja wielomianowa (Stopień {})'.format(degree))

plt.tight_layout()
plt.show()
```
