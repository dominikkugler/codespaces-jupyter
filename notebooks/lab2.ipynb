```python
import numpy as np
def f(x):
    try:
        return 1/x
    except ZeroDivisionError: print("blad: dzielenie przez 0")  
def suma(funkcja,xp,xk,krok):
    wartosc = 0
    for i in np.arange(xp,xk,krok):
        print(i)
        wartosc += funkcja(i)
    print(wartosc)
def iloczyn(funkcja,xp,xk,krok):
    wartosc = 1
    for i in np.arange(xp,xk,krok):
        print(i)
        wartosc *= funkcja(i)
    print(wartosc)   

suma(np.sin,-3.14,3.14,0.01)
suma(f,-3.14,3.14,0.01)
iloczyn(f,-3.14,3.14,0.01)
iloczyn(np.sin,-3.14,3.14,0.01)
```

```python
import matplotlib.pyplot as plt 
import numpy as np
def rysuj(funkcja, xp, xk, krok,opcja):
    x = [i for i in np.arange(xp,xk,krok)]
    y = [funkcja(i) for i in np.arange(xp,xk,krok)]
    if (opcja==1):
        plt.plot(x,y, 'g.')
        plt.ylabel('f(x)')
        plt.xlabel('x')
        plt.grid(visible=True, axis='both')
        plt.show
    else:
        plt.plot(x,y)
        plt.ylabel('f(x)')
        plt.xlabel('x')
        plt.show

rysuj(np.sin,0,6,0.2,1)
```
