```python
import numpy as np
import scipy as sp
import timeit as t
def fx1(x):
    return 7*x**5 - 6*x**2 - 2*x

def fx2(x):
    return 1/((x-0.3)**2+0.01) - 1/((x-0.8)**2+0.04)

def fx3(x):
    return x**4 - 6.4*x**3 + 6.45*x**2 + 20.538*x - 31.752

#Zad.1
def secant(fx,a,b,err):
    fb=fx(b)
    while np.absolute(fb)>err:
        midPoint = b - (b-a)*fb/(fb-fx(a))
        a=b
        b=midPoint
        fb=fx(b)
    return b

def newton_raphson(fx,x0,err):
    x=x0
    h=0.1e-5
    while np.absolute(fx(x))>err:
        d1fx = (fx(x+h/2.0)-fx(x-h/2.0))/h
        x=x-fx(x)/d1fx
    return x

def newt(f):
    return sp.optimize.newton(f,0.2)
def bis(f):
    return sp.optimize.bisect(f,0,1)
def ridd(f):
    return sp.optimize.ridder(f,0,1)
def bren(f):
    return sp.optimize.brenth(f,0,1)

if __name__ == '__main__':
    print("Metoda siecznych")
    print("Funkcja 1")
    x1f1 = secant(fx1,-0.5,-0.25,0.0001)
    err_x1f1 = fx1(x1f1)
    print("x1 = ",x1f1)
    print("err = ",err_x1f1)
    x2f1 = secant(fx1,-0.15,0.15,0.0001)
    err_x2f1 = fx1(x2f1)
    print("x2 = ",x2f1)
    print("err = ",err_x2f1)
    x3f1 = secant(fx1,1,1.25,0.0001)
    err_x3f1 = fx1(x3f1)
    print("x3 = ",x3f1)
    print("err = ",err_x3f1)

    print("Funkcja 2")
    x1f2 = secant(fx2,0.25,0.75,0.0001)
    err_x1f2 = fx2(x1f2)
    print("x1 = ",x1f2)
    print("err = ",err_x1f2)

    print("Metoda Newtona-Raphsona")
    print("Funkcja 1")
    nrx1f1 = newton_raphson(fx1,-0.5,0.0001)
    err_nrx1f1 = fx1(nrx1f1)
    print("x1 = ",nrx1f1)
    print("err = ",err_nrx1f1)
    nrx2f1 = newton_raphson(fx1,-0.15,0.0001)
    err_nrx2f1 = fx1(nrx2f1)
    print("x2 = ",nrx2f1)
    print("err = ",err_nrx2f1)
    nrx3f1 = newton_raphson(fx1,1,0.0001)
    err_nrx3f1 = fx1(nrx3f1)
    print("x3 = ",nrx3f1)
    print("err = ",err_nrx3f1)

    print("Funkcja 2")
    nrx1f2 = newton_raphson(fx2,0.50,0.0001)
    err_nrx1f2 = fx2(nrx1f2)
    print("x1 = ",nrx1f2)
    print("err = ",err_nrx1f2)

    print("Funkcja 3")
    nrx1f3 = newton_raphson(fx3,-2,0.0001)
    err_nrx1f3 = fx3(nrx1f3)
    print("x1 = ",nrx1f3)
    print("err = ",err_nrx1f3)
    nrx2f3 = newton_raphson(fx3,2,0.0001)
    err_nrx2f3 = fx3(nrx2f3)
    print("x2 = ",nrx2f3)
    print("err = ",err_nrx2f3)
    nrx3f3 = newton_raphson(fx3,3.5,0.0001)
    err_nrx3f3 = fx3(nrx3f3)
    print("x3 = ",nrx3f3)
    print("err = ",err_nrx3f3)

    print("Test metody bisekcji -funkcja biblioteczna")
    print("Funkcja 1")
    print("x1 = ",sp.optimize.bisect(fx1,-0.5,-0.25))
    print("x2 = ",sp.optimize.bisect(fx1,-0.15,0.15))
    print("x3 = ",sp.optimize.bisect(fx1,1,2))

    print("Funkcja 2")
    print("x1 = ",sp.optimize.bisect(fx2,0.25,0.75))

    print("Test metody Newtona-Raphsona -funkcja biblioteczna")
    print("Funkcja 1")
    print("x1 = ",sp.optimize.newton(fx1,-0.5))
    print("x2 = ",sp.optimize.newton(fx1,-0.15))
    print("x3 = ",sp.optimize.newton(fx1,1))

    print("Funkcja 3")
    print("x1 = ",sp.optimize.newton(fx3,-2))
    print("x2 = ",sp.optimize.newton(fx3,2))
    print("x3 = ",sp.optimize.newton(fx3,3.5))

    print("Test metody Riddera -funkcja biblioteczna")
    print("Funkcja 1")
    print("x1 = ",sp.optimize.ridder(fx1,-0.5,-0.25))
    print("x2 = ",sp.optimize.ridder(fx1,-0.15,0.15))
    print("x3 = ",sp.optimize.ridder(fx1,1,2))

    print("Funkcja 2")
    print("x1 = ",sp.optimize.ridder(fx2,0.25,0.75))

    print("Test metody Brenth -funkcja biblioteczna")
    print("Funkcja 1")
    print("x1 = ",sp.optimize.brenth(fx1,-0.5,-0.25))
    print("x2 = ",sp.optimize.brenth(fx1,-0.15,0.15))
    print("x3 = ",sp.optimize.brenth(fx1,1,2))

    print("Funkcja 2")
    print("x1 = ",sp.optimize.brenth(fx2,0.25,0.75))

    print("Czasy wykonania")
    print("Metoda bisekcji - funkcja wbudowana")
    print(t.timeit('bis(fx1)',number=1000,setup='from __main__ import fx1,bis'))
    print(t.timeit('bis(fx2)',number=1000,setup='from __main__ import fx2,bis'))
    print("Metoda Newtona-Raphsona - funkcja wbudowana")
    print(t.timeit('newt(fx1)',number=1000,setup='from __main__ import fx1,newt'))
    print(t.timeit('newt(fx3)',number=1000,setup='from __main__ import fx3,newt'))
    print("Metoda Riddera - funkcja wbudowana")
    print(t.timeit('ridd(fx1)',number=1000,setup='from __main__ import fx1,ridd'))
    print(t.timeit('ridd(fx2)',number=1000,setup='from __main__ import fx2,ridd'))
    print("Metoda Brenth - funkcja wbudowana")
    print(t.timeit('bren(fx1)',number=1000,setup='from __main__ import fx1,bren'))
    print(t.timeit('bren(fx2)',number=1000,setup='from __main__ import fx2,bren'))

    #Wnioski
    # Z wszystkich wbudowanych metod z biblioteki SciPy metoda Newtona-Raphsona
    # jest kilkukrotnie wolniejsza od pozostałych metod.
    # Metody wbudowane z biblioteki SciPy są dokładniejsze od metod własnoręcznie napisanych
    # Metody własnoręcznie napisane są bardzo dokładne, generują błedy rzędu 10^-5







```
