```python
def interplin2p(x,xi,yi,xi1,yi1):
    return yi+(x-xi)*(yi1-yi)/(xi1-xi)
def interlinevect(x,xyvect):
    xinterp = []
    yinterp = []
    for xk in x:
        N=len(xyvect[0])
        for i in range(0,N-1):
            if(xk>=xyvect[0][i] and xk<xyvect[0][i+1]):
                xinterp.append(xk)
                yinterp.append(interplin2p(xk,xyvect[0][i],xyvect[1][i],xyvect[0][i+1],xyvect[1][i+1]))
            i=i+1
    return [xinterp,yinterp]
```

```python
def interpolacja_lagrange(x,y,xval):
    products=0
    yval=0
    for i in range(len(x)):
        products=y[i]
        for j in range(len(x)):
            if i!=j:
                products=products*(xval-x[j])/(x[i]-x[j])
        yval=yval+products
    return yval
```

```python
import numpy as np
xp = 0
xk = 2*np.pi
krok=0.1
liczbaPunktow = np.abs((xk-xp)/krok)
liczbaPunktowInt = int(np.ceil(liczbaPunktow))
print("Liczba punktow wartosci to %d" % liczbaPunktow)
x = xp
for i in range(0, liczbaPunktowInt):
    x=x+krok
    print("sin(%f)=%f" % (x, np.sin(x)))
    print("cos(%f)=%f" % (x,np.cos(x)))
    print("tan(%f)=%f" % (x,np.tan(x)))
    print("ctg(%f)=%f" % (x,1.0/np.sin(x)))
```
