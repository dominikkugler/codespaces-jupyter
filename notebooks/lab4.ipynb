```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
def interplin2p(x,xi,yi,xi1,yi1):
    return yi+(x-xi)*(yi1-yi)/(xi1-xi)
def interlinevect(x,xyvect):
    xinterp = []
    yinterp = []
    for xk in x:
        N=len(xyvect[0])
        for i in range(0,N-1):
            if(xk>=xyvect[0][i] and xk<xyvect[0][i+1]):
                xinterp.append(xk)
                yinterp.append(interplin2p(xk,xyvect[0][i],xyvect[1][i],xyvect[0][i+1],xyvect[1][i+1]))
            i=i+1
    return [xinterp,yinterp]

xyv = []
xyv.append([i for i in np.arange(0,3,0.1)])
xyv.append([np.cos(1/i) for i in np.arange(0,3,0.1)])
print(xyv)
print(len(xyv[0]))
pylab.plot(xyv[0],xyv[1],'.', label='original')

xin = [i for i in np.arange(0,3,0.08)]
xyinterp = interlinevect(xin,xyv)
print(xyinterp)
pylab.plot(xyinterp[0],xyinterp[1],'^-r', label='interpolated')
pylab.title('cos(1/x)')
pylab.legend()
pylab.show()
print(len(xyinterp[0]))
print(len(xyinterp[1]))
print(len(xin))


xyexp = []
xyexp.append([i for i in np.arange(0,3,0.1)])
xyexp.append([np.exp(np.sin(i)) for i in np.arange(0,3,0.1)])
print(xyexp)
print(len(xyexp[0]))
pylab.plot(xyexp[0],xyexp[1],'.', label='original')

xin_exp = [i for i in np.arange(0,3,0.08)]
xyinterp_exp = interlinevect(xin_exp,xyexp)
print(xyinterp_exp)
pylab.plot(xyinterp_exp[0],xyinterp_exp[1],'^-r', label='interpolated')
pylab.title('exp(sin(x))')
pylab.legend()
pylab.show()
print(len(xyinterp_exp[0]))
print(len(xyinterp_exp[1]))
print(len(xin_exp))
```

```python
import pylab
import numpy as np

def interpolacja_lagrange(x,y,xval):
    products=0
    yval=0
    for i in range(len(x)):
        products=y[i]
        for j in range(len(x)):
            if i!=j:
                products=products*(xval-x[j])/(x[i]-x[j])
        yval=yval+products
    return yval

x_sin=[i for i in np.arange(0,3,0.1)]
y_sin=[np.sin(1/i) for i in np.arange(0,3,0.1)]

pylab.plot(x_sin,y_sin,'.', label='original')
xval_sin=[i for i in np.arange(0,3,0.08)]
yval_sin=[]
for xv_sin in xval_sin:
    yval_sin.append(interpolacja_lagrange(x_sin,y_sin,xv_sin))
pylab.plot(xval_sin,yval_sin,'-', label='interpolated')
pylab.title('sin(1/x)')
pylab.legend()
pylab.show()

x_exp=[i for i in np.arange(0,3,0.1)]
y_exp=[np.exp(np.cos(i)) for i in np.arange(0,3,0.1)]

pylab.plot(x_exp,y_exp,'.', label='original')
xval_exp=[i for i in np.arange(0,3,0.08)]
yval_exp=[]
for xv_exp in xval_exp:
    yval_exp.append(interpolacja_lagrange(x_exp,y_exp,xv_exp))
pylab.plot(xval_exp,yval_exp,'-', label='interpolated')
pylab.title('exp(cos(x))')
pylab.legend()
pylab.show()
```
